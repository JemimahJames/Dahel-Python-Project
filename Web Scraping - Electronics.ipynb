```python
from bs4 import BeautifulSoup 
import requests 
```

```python
link = 'https://jiji.ng/electronics'
```

```python
requests.get(link)
```

```python
req = requests.get(link)
```

```python
swallow = BeautifulSoup(req.text, 'html')
```

```python
print(swallow)
```

```python
swallow.find_all('div', class_ ='b-list-advert-base__data')
```

```python
price = swallow.select("div.qa-advert-price")
```

```python
prices = [i.text.strip() for i in price]
print(prices)
```

```python
swallow.select("div.b-advert-title-inner.qa-advert-title.b-advert-title-inner--div")
```

```python
name =swallow.select("div.b-advert-title-inner.qa-advert-title.b-advert-title-inner--div")
```

```python
names = [i.text.strip() for i in name]
print(names)
```

```python
swallow.select("div.b-list-advert-base__description-text")
```

```python
description = swallow.select("div.b-list-advert-base__description-text")
```

```python
descriptions = [i.text.strip() for i in description]
print(descriptions)
```

```python
import pandas as pd
```

```python
table = pd.DataFrame({
    'name': names,
    'price': prices,
    'description': descriptions
})

print(table)
```
